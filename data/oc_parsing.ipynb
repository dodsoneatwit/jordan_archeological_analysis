```python
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
import json as js
```

```python
path = 'coin_artifacts.json'
```

```python

with open(path, 'r') as json:
   data = js.loads(json.read())
```

```python
norm = json_normalize(data['features'])
norm.head()
```

```python
clean = norm[norm['properties.thumbnail'].notnull()]
```

```python
remove_col = clean.drop(columns=['count', 'type', 'category', 'when.id','when.start','when.stop', 'when.type','geometry.id','json',
                                'id', 'properties.label','properties.feature-type', 'properties.id',
                                'properties.count','properties.href', 'rdfs:isDefinedBy', 'properties.uri', 'properties.citation uri',
                                'properties.project href', 'geometry.type', 'properties.published', 'properties.updated',
                                'properties.context href'])
remove_col
```

```python
remove_col.to_csv('oc_parsed_artifacts.csv')
```
